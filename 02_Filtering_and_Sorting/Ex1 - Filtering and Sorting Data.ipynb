```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

chipo = pd.read_csv(url, sep = '\t')
```

```python
chipo.head()
```

```python
chipo['num_price'] = chipo.item_price.str.slice(1).astype(float)
```

```python
a1 = chipo.groupby('item_name').sum()
a1['price'] = (a1.num_price)/(a1.quantity)
a1[a1['price']>10]
```
